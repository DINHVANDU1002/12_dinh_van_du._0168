```python
class NhanVien:
    def __init__(self, ma_nv, luong_cb):
        self.__ma_nv = ma_nv        # thuộc tính private
        self.__luong_cb = luong_cb  # thuộc tính private

    def tinh_luong(self):
        return self.__luong_cb

    def get_ma_nv(self):
        return self.__ma_nv

    def get_luong_cb(self):
        return self.__luong_cb
```

```python
class NhanVienTangCa(NhanVien):
    def tinh_luong(self):
        return self.get_luong_cb() * 1.2
```

```python
class NhanVienChuyenChuyen(NhanVien):
    def tinh_luong(self):
        return self.get_luong_cb() + 500000
```

```python
if __name__ == "__main__":
    nv1 = NhanVien("NV01", 10000000)
    nv2 = NhanVienTangCa("NV02", 10000000)
    nv3 = NhanVienChuyenChuyen("NV03", 10000000)

    danh_sach = [nv1, nv2, nv3]

    for nv in danh_sach:
        print("Mã NV:", nv.get_ma_nv(), 
              "- Lương:", nv.tinh_luong())
```

```python
san_luong = [12, None, 8, 15, None, 10]
```

```python
import numpy as np

arr = np.array(san_luong, dtype=float)
arr = arr.reshape(2, 3)
```

```python
mean_value = np.nanmean(arr)
arr = np.nan_to_num(arr, nan=mean_value)
```

```python
lon_hon_9 = arr[arr > 9]
gia_tri_nho_nhat = np.min(lon_hon_9)

print("Các phần tử > 9:", lon_hon_9)
print("Giá trị nhỏ nhất:", gia_tri_nho_nhat)
```
