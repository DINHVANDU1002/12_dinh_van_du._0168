```python
import pandas as pd
import sqlite3

df = pd.read_csv("nhan_su.csv")
```

```python
df_loc = df[
    df["Ma_NV"].str.contains("NS", case=False, na=False) &
    (df["Nam_Cong_Tac"] >= 2021)
]
```

```python
mean_thanh_tich = df_loc["Thanh_Tich"].mean()
df_loc["Thanh_Tich"] = df_loc["Thanh_Tich"].fillna(mean_thanh_tich)

df_loc
```

```python
tong_thanh_tich = df_loc.groupby("Phong_Ban")["Thanh_Tich"].sum()
```

```python
pivot_table = pd.pivot_table(
    df_loc,
    index="Phong_Ban",
    columns="Nam_Cong_Tac",
    values="Thanh_Tich",
    aggfunc="sum",
    fill_value=0
)
```

```python
conn = sqlite3.connect("HeThong.db")

pivot_table.to_sql(
    "bang_luong_2024",
    conn,
    if_exists="replace"
)
```
